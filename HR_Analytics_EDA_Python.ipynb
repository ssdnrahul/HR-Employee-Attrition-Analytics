```python
import pandas as pd
pd.set_option('display.max_columns', None)
df=pd.read_csv("Employees.csv")
```

```python
#Which department has the highest attrition?
department_count = df[df["Attrition"]=="Yes"]["Department"].value_counts()
total_employees = df["Department"].value_counts()
attrition_per = round((department_count / total_employees) * 100,2).sort_values(ascending=False)
attrition_per
```

```python
# Which Job Role Has the Highest Attrition?

jobrole_count = df[df["Attrition"]=="Yes"]["JobRole"].value_counts()
total_jobrole = df["JobRole"].value_counts()
jobrole_attrition_per = round((jobrole_count / total_jobrole) * 100,2).sort_values(ascending=False)
jobrole_attrition_per
```

```python
#What percentage of employees working overtime leave the company?
overtime_count = df[df["Attrition"]=="Yes"]["OverTime"].value_counts()
total_overtime=df["OverTime"].value_counts()
overtime_attrition_per = round((overtime_count / total_overtime)*100,2).sort_values(ascending=False)
overtime_attrition_per
```

```python
# Which age group has the highest attrition rate?

def age_group(age):
    if age<=30:
        return "18-30"
    elif age<=40:
        return "31-40"
    elif age<=50:
        return "41-50"
    else:
        return "51-60"
df["Age_Group"]=df["Age"].apply(
    age_group
)
df

agegroupcount=df[df["Attrition"]=="Yes"]["Age_Group"].value_counts()
total_agegroup=df["Age_Group"].value_counts()
agegroup_attrition_per= round((agegroupcount / total_agegroup)*100,2).sort_values(ascending=False)
agegroup_attrition_per
```

```python
# What is the relationship between salary and attrition?

def salary_group(salary):
    if salary < 5000:
        return "Low"
    elif salary < 10000:
        return "Medium"
    else: 
        return "High"
df["Salary_Group"]=df["MonthlyIncome"].apply(
    salary_group
)
df

salary_count = df[df["Attrition"]=="Yes"]["Salary_Group"].value_counts()
total_salary = df["Salary_Group"].value_counts()
salary_attrition_per = round((salary_count / total_salary)*100,2).sort_values(ascending=False)
salary_attrition_per
```

```python
# Does a lack of promotion increase employee attrition?
def promotion_group(years):

    if years <= 3:
        return "0-3 Years"

    elif years <= 7:
        return "4-7 Years"

    elif years <= 11:
        return "8-11 Years"
    else:
        return "12+ Years"

df["Promotion_Group"] = df["YearsSinceLastPromotion"].apply(promotion_group)
df

promotion_count = df[df["Attrition"]=="Yes"]["Promotion_Group"].value_counts()

total_promotion = df["Promotion_Group"].value_counts()

promotion_attrition_per = round(
    (promotion_count / total_promotion) * 100,
    2
).sort_values(ascending=False)

promotion_attrition_per
```

```python
# Which group has higher attrition: Single or Married employees?

marital_count = df[df["Attrition"]=="Yes"]["MaritalStatus"].value_counts()

total_marital = df["MaritalStatus"].value_counts()

marital_attrition_per = round((marital_count / total_marital) * 100,2).sort_values(ascending=False)

marital_attrition_per
```

```python
#Does Work-Life Balance affect employee attrition?
wlb_count = df[df["Attrition"]=="Yes"]["WorkLifeBalance"].value_counts()

total_wlb = df["WorkLifeBalance"].value_counts()

wlb_attrition_per = round((wlb_count / total_wlb) * 100,2).sort_index()

wlb_attrition_per
```
